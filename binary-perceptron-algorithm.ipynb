```python
'''

Implementation of a basic binary perceptron with the most basic version of the algorithm (without using pre-defined packages or libraries).
We use the binary perceptron to train classifiers to discriminate between
• class 1 and class 2,
• class 2 and class 3, and
• class 1 and class 3.

And reporting the train and test classification accuracies for each of the three classifiers after training for 20 iterations. 

Extending the binary perceptron implemented earlier, to perform multi-class classification
using the 1-vs-rest approach. Reporting the train and test classification accuracies for the
multi-class classifier after training for 20 iterations.

Lastly, adding a L2 regularisation term to the multi-class classifier implemented above with the regularisation coefficient set to : 0.01, 0.1, 1.0, 10.0, 100.0.

'''
```

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

# --------------- TRAIN DATA ---------------
# Using pandas to read the training data
train_data = pd.read_csv('/kaggle/input/perceptron-train-and-test-data/train.data', sep=",", header=None)

# Converting the dataframe into array for Binary Perceptron
temp12 = np.array(train_data[train_data[4]!='class-3']) # Filtering data points only with class 1 & 2
temp23 = np.array(train_data[train_data[4]!='class-1']) # Filtering data points only with class 2 & 3
temp13 = np.array(train_data[train_data[4]!='class-2']) # Filtering data points only with class 1 & 3

# Choosing the first 4 columns of the array as the training dataset X for the 3 created arrays
X_train12 = temp12[:,:4]
n = np.shape(X_train12)[0]
X_train12 = np.reshape(X_train12, (n,4))

X_train23 = temp23[:,:4]
n = np.shape(X_train23)[0]
X_train23 = np.reshape(X_train23, (n,4))

X_train13 = temp13[:,:4]
n = np.shape(X_train13)[0]
X_train13 = np.reshape(X_train13, (n,4))

# Preparing X train data for One Vs Rest approach with all class labels
temp = np.array(train_data)
X_train = temp[:,:4]
n = np.shape(X_train)[0]
X_train = np.reshape(X_train, (n,4))

# Similarly choosing the last column of the dataset of class labels and replacing class 1 with value 1 and
# class 2 with value -1
y_train12 = temp12[:,4]
n = np.shape(y_train12)[0]
y_train12 = np.reshape(y_train12, (n,1))
y_train12[y_train12 == 'class-1'] = 1
y_train12[y_train12 == 'class-2'] = -1

# Class 2 is 1 and class 3 is -1
y_train23 = temp23[:,4]
n = np.shape(y_train23)[0]
y_train23 = np.reshape(y_train23, (n,1))
y_train23[y_train23 == 'class-2'] = 1
y_train23[y_train23 == 'class-3'] = -1

# Class 1 is 1 and class 3 is -1
y_train13 = temp13[:,4]
n = np.shape(y_train13)[0]
y_train13 = np.reshape(y_train13, (n,1))
y_train13[y_train13 == 'class-1'] = 1
y_train13[y_train13 == 'class-3'] = -1

# Preparing y test data for One Vs Rest approach with all class labels
temp1 = np.array(train_data)
temp2 = np.array(train_data)
temp3 = np.array(train_data)

# Class 1 is 1 and class 2 & 3 is -1
y_train1 = temp1[:,4]
n = np.shape(y_train1)[0]
y_train1 = np.reshape(y_train1, (n,1))
y_train1[y_train1 == 'class-1'] = 1
y_train1[y_train1 == 'class-2'] = -1
y_train1[y_train1 == 'class-3'] = -1

# Class 2 is 1 and class 1 & 3 is -1
y_train2 = temp2[:,4]
n = np.shape(y_train2)[0]
y_train2 = np.reshape(y_train2, (n,1))
y_train2[y_train2 == 'class-1'] = -1
y_train2[y_train2 == 'class-2'] = 1
y_train2[y_train2 == 'class-3'] = -1

# Class 3 is 1 and class 1 & 2 is -1
y_train3 = temp3[:,4]
n = np.shape(y_train3)[0]
y_train3 = np.reshape(y_train3, (n,1))
y_train3[y_train3 == 'class-1'] = -1
y_train3[y_train3 == 'class-2'] = -1
y_train3[y_train3 == 'class-3'] = 1

# --------------- TEST DATA ---------------
# Using pandas to read the testing data
test_data = pd.read_csv('/kaggle/input/perceptron-train-and-test-data/test.data', sep=",", header=None)

# Converting the dataframe into array for Binary Perceptron
temp12 = np.array(test_data[test_data[4]!='class-3']) # Filtering data points only with class 1 & 2
temp23 = np.array(test_data[test_data[4]!='class-1']) # Filtering data points only with class 2 & 3
temp13 = np.array(test_data[test_data[4]!='class-2']) # Filtering data points only with class 1 & 3

# Choosing the first 4 columns of the array as the testing dataset X for the 3 created arrays
X_test12 = temp12[:,:4]
n = np.shape(X_test12)[0]
X_test12 = np.reshape(X_test12, (n,4))

X_test23 = temp23[:,:4]
n = np.shape(X_test23)[0]
X_test23 = np.reshape(X_test23, (n,4))

X_test13 = temp13[:,:4]
n = np.shape(X_test13)[0]
X_test13 = np.reshape(X_test13, (n,4))

# Preparing X test data for One Vs Rest approach with all class labels
temp = np.array(test_data)
X_test = temp[:,:4]
n = np.shape(X_test)[0]
X_test = np.reshape(X_test, (n,4))

# Similarly choosing the last column of the dataset of class labels and replacing class 1 with value 1 and
# class 2 with value -1
y_test12 = temp12[:,4]
n = np.shape(y_test12)[0]
y_test12 = np.reshape(y_test12, (n,1))
y_test12[y_test12 == 'class-1'] = 1
y_test12[y_test12 == 'class-2'] = -1

# Class 2 is 1 and class 3 is -1
y_test23 = temp23[:,4]
n = np.shape(y_test23)[0]
y_test23 = np.reshape(y_test23, (n,1))
y_test23[y_test23 == 'class-2'] = 1
y_test23[y_test23 == 'class-3'] = -1

# Class 1 is 1 and class 3 is -1
y_test13 = temp13[:,4]
n = np.shape(y_test13)[0]
y_test13 = np.reshape(y_test13, (n,1))
y_test13[y_test13 == 'class-1'] = 1
y_test13[y_test13 == 'class-3'] = -1

# Preparing y test data for One Vs Rest approach with all class labels
# Class 1 is 1 and class 2 & 3 is -1
y_test1 = temp[:,4]
n = np.shape(y_test1)[0]
y_test1 = np.reshape(y_test1, (n,1))
y_test1[y_test1 == 'class-1'] = 1
y_test1[y_test1 == 'class-2'] = -1
y_test1[y_test1 == 'class-3'] = -1

# Class 2 is 1 and class 1 & 3 is -1
y_test2 = temp[:,4]
n = np.shape(y_test2)[0]
y_test2 = np.reshape(y_test2, (n,1))
y_test2[y_test2 == 'class-1'] = -1
y_test2[y_test2 == 'class-2'] = 1
y_test2[y_test2 == 'class-3'] = -1

# Class 3 is 1 and class 1 & 2 is -1
y_test3 = temp[:,4]
n = np.shape(y_test3)[0]
y_test3 = np.reshape(y_test3, (n,1))
y_test3[y_test3 == 'class-1'] = -1
y_test3[y_test3 == 'class-2'] = -1
y_test3[y_test3 == 'class-3'] = 1

# Binary Perceptron Training Function
def train_perceptron(X_train, y_train):
    # Initializing weights and bias
    w = [0,0,0,0]
    b = 0

    # Setting the learning rate and number of iterations
    lr = 1
    n_iter = 20

    # Training the perceptron
    for i in range(n_iter):
        for j in range(np.shape(X_train)[0]):
            x = X_train[j]
            y_true = y_train[j]
            y_pred = np.sign(np.dot(x, w) + b)
            if y_pred != y_true:
                w += lr * y_true * x
                b += lr * y_true
    return w,b

# Multiclass Perceptron Implementation with L2 regularisation coefficient
def train_perceptron_reg(X_train, y_train, reg):
    w = [0,0,0,0]
    b = 0
    lr = 1
    n_iter = 20

    for i in range(n_iter):
        for j in range(np.shape(X_train)[0]):
            x = X_train[j]
            y_true = y_train[j]
            y_pred = np.sign(np.dot(x, w) + b)
            if y_pred != y_true:
                w += lr * y_true * x - 2 * lr * np.dot(reg, w) # Including the regularisation coefficient
                b += lr * y_true
    return w,b

# Calculating the predicted output using test data
def predict(X, weight, bias):
    y_pred = np.sign(np.dot(X, weight) + bias)
    return y_pred

# Accuracy Calculation Function
def accuracy(y_true, y_pred):
    y_pred = np.reshape(y_pred, (np.shape(y_pred)[0], 1))
    accuracy = (np.sum(np.equal(y_true, y_pred))/len(y_true)) * 100
    return accuracy

# Training the Binary Perceptron to produce the predicted values lastly to produce the accuracy
w,b = train_perceptron(X_train12, y_train12)
y_pred12 = predict(X_train12, w, b)
acc12 = accuracy(y_train12, y_pred12)

w,b = train_perceptron(X_train23, y_train23)
y_pred23 = predict(X_train23, w, b)
acc23 = accuracy(y_train23, y_pred23)

w,b = train_perceptron(X_train13, y_train13)
y_pred13 = predict(X_train13, w, b)
acc13 = accuracy(y_train13, y_pred13)

print("\n======== IMPLEMENTATION OF BINARY PERCEPTRON - TRAINING ACCURACIES ========")
print("Accuracy for class 1 and class 2:")
print("%.2f" %acc12, "%")
print("Accuracy for class 2 and class 3:")
print("%.2f" %acc23, "%")
print("Accuracy for class 1 and class 3:")
print("%.2f" %acc13, "%")

w,b = train_perceptron(X_train12, y_train12)
y_pred12 = predict(X_test12, w, b)
acc12 = accuracy(y_test12, y_pred12)

w,b = train_perceptron(X_train23, y_train23)
y_pred23 = predict(X_test23, w, b)
acc23 = accuracy(y_test23, y_pred23)

w,b = train_perceptron(X_train13, y_train13)
y_pred13 = predict(X_test13, w, b)
acc13 = accuracy(y_test13, y_pred13)

print("\n======== IMPLEMENTATION OF BINARY PERCEPTRON - TESTING ACCURACIES ========")
print("Accuracy for class 1 and class 2:")
print("%.2f" %acc12, "%")
print("Accuracy for class 2 and class 3:")
print("%.2f" %acc23, "%")
print("Accuracy for class 1 and class 3:")
print("%.2f" %acc13, "%")

# Training the One vs Rest Approach based Perceptron to produce the predicted values lastly to produce the accuracy
w,b = train_perceptron(X_train, y_train1)
y_pred1 = predict(X_train, w, b)
acc1 = accuracy(y_train1, y_pred1)

w,b = train_perceptron(X_train, y_train2)
y_pred2 = predict(X_train, w, b)
acc2 = accuracy(y_train2, y_pred2)

w,b = train_perceptron(X_train, y_train3)
y_pred3 = predict(X_train, w, b)
acc3 = accuracy(y_train3, y_pred3)

print("\n======== IMPLEMENTATION OF ONE VS REST APPROACH IN PERCEPTRON - TRAINING ACCURACIES ========")
print("Accuracy for class 1 :")
print("%.2f" %acc1, "%")
print("Accuracy for class 2 :")
print("%.2f" %acc2, "%")
print("Accuracy for class 3 :")
print("%.2f" %acc3, "%")

w,b = train_perceptron(X_train, y_train1)
y_pred1 = predict(X_test, w, b)
acc1 = accuracy(y_test1, y_pred1)

w,b = train_perceptron(X_train, y_train2)
y_pred2 = predict(X_test, w, b)
acc2 = accuracy(y_test2, y_pred2)

w,b = train_perceptron(X_train, y_train3)
y_pred3 = predict(X_test, w, b)
acc3 = accuracy(y_test3, y_pred3)

print("\n======== IMPLEMENTATION OF ONE VS REST APPROACH IN PERCEPTRON - TESTING ACCURACIES ========")
print("Accuracy for class 1 :")
print("%.2f" %acc1, "%")
print("Accuracy for class 2 :")
print("%.2f" %acc2, "%")
print("Accuracy for class 3 :")
print("%.2f" %acc3, "%")

# Training the One vs Rest Approach based Perceptron with regularisation coefficients
# to produce the predicted values lastly to produce the accuracy
reg_coeff = [0.01, 0.1, 1, 10, 100]

for i in range(3):
    w, b = train_perceptron_reg(X_train, y_train1, reg_coeff[i])
    y_pred1 = predict(X_train, w, b)
    acc1 = accuracy(y_train1, y_pred1)

    w, b = train_perceptron_reg(X_train, y_train2, reg_coeff[i])
    y_pred2 = predict(X_train, w, b)
    acc2 = accuracy(y_train2, y_pred2)

    w, b = train_perceptron_reg(X_train, y_train3, reg_coeff[i])
    y_pred3 = predict(X_train, w, b)
    acc3 = accuracy(y_train3, y_pred3)

    print(f"\n======== IMPLEMENTATION OF L2 REGULARISATION OF {reg_coeff[i]} IN PERCEPTRON - TRAINING ACCURACIES ========")
    print("Accuracy for class 1 :")
    print("%.2f" %acc1, "%")
    print("Accuracy for class 2 :")
    print("%.2f" %acc2, "%")
    print("Accuracy for class 3 :")
    print("%.2f" %acc3, "%")

for i in range(3):
    w, b = train_perceptron_reg(X_train, y_train1, reg_coeff[i])
    y_pred1 = predict(X_test, w, b)
    acc1 = accuracy(y_test1, y_pred1)

    w, b = train_perceptron_reg(X_train, y_train2, reg_coeff[i])
    y_pred2 = predict(X_test, w, b)
    acc2 = accuracy(y_test2, y_pred2)

    w, b = train_perceptron_reg(X_train, y_train3, reg_coeff[i])
    y_pred3 = predict(X_test, w, b)
    acc3 = accuracy(y_test3, y_pred3)

    print(f"\n======== IMPLEMENTATION OF L2 REGULARISATION OF {reg_coeff[i]} IN PERCEPTRON - TESTING ACCURACIES ========")
    print("Accuracy for class 1 :")
    print("%.2f" %acc1, "%")
    print("Accuracy for class 2 :")
    print("%.2f" %acc2, "%")
    print("Accuracy for class 3 :")
    print("%.2f" %acc3, "%")
```
